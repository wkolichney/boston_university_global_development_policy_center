```python
import pandas as pd
```

```python
bu = pd.read_csv('C:/Users/wikku/global_china_initiative/workspace/aiddata_join/GCI Database-Loans.csv')
aiddata = pd.read_excel('C:/Users/wikku/global_china_initiative/workspace/aiddata_join/AidDatas_CLG_Global_Dataset_v1.0.xlsx', sheet_name='CLG-Global 1.0_Records')
```

```python
bu_concise = bu[['ID', 'BU ID', 'AidData ID']]
bu_concise.rename(columns={'AidData ID': 'AidData_Record_ID'}, inplace=True)
bu_concise = bu_concise[bu_concise['AidData_Record_ID'] != "No Match with '23 Cross Check"]
bu_concise = bu_concise[bu_concise['AidData_Record_ID'] != "Associated with, not exact match for: 61056"]
bu_concise['AidData_Record_ID'] = bu_concise['AidData_Record_ID'].str.strip(' ;')
bu_concise['AidData_Record_ID'] = bu_concise['AidData_Record_ID'].str.replace(' ', '', regex=False)
bu_expanded = bu_concise.assign(
    AidData_Record_ID=bu_concise['AidData_Record_ID'].str.split(r'[,;&]+', regex=True)
).explode('AidData_Record_ID')
bu_expanded = bu_expanded[bu_expanded['AidData_Record_ID'] != '']
bu_expanded = bu_expanded[bu_expanded['AidData_Record_ID'] != "c"]
bu_expanded = bu_expanded[bu_expanded['AidData_Record_ID'].notna()]
bu_expanded["AidData_Record_ID"] = bu_expanded["AidData_Record_ID"].astype(int)
```

```python
join = pd.merge(bu_expanded, aiddata, on = "AidData_Record_ID", how='left')
```

```python
join.to_excel('C:/Users/wikku/global_china_initiative/workspace/aiddata_join/aiddata_bu_joined.xlsx', index=False)
```
